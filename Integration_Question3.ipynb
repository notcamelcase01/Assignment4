## Question 3


$x \in [a,b]$

Number of Intervals : $n$

Interval Size : $\frac{b-a}{n}$

$I = \int_a^b\sin xdx$

Exact solution of the above  integral is $2$

```python
import numpy as np
 


def WriteDataToFile(list,myfile):
    myfile.write(', '.join(str(item) for item in list)+'\n')



F = lambda x: np.sin(x)

I_exact = 2.
xb = np.pi
xa = 0.

def TrapezoidalRuleIntegration(n):
    x = np.linspace(xa,xb,n+1) ##(n + 1)Plus one because linspace creates n point we want n intervals
    s=0
    [s:=s+2*F(i) for i in x]
    return (s-F(xb)-F(xa))*0.5*(xb-xa)/n

def SimpsonRuleIntegration(n): 
    x = np.linspace(xa,xb,n+1) ##(n + 1)Plus one because linspace creates n point we want n intervals
    s=F(xa)+F(xb)
    [s:=s+4*F(x[i]) for i in range(1,len(x)-1,2)]
    [s:=s+2*F(x[i]) for i in range(2,len(x)-1,2)]
    return s*(xb-xa)/n/3
    
    

    
myfile = open("output_data\\trap.csv", 'a') 
myfile.seek(0) 
myfile.truncate() 
WriteDataToFile(["n","h","EstimateIntegral","eh","e2h/eh"],myfile)

firstiter = TrapezoidalRuleIntegration(1)
WriteDataToFile([1,(xb-xa),firstiter,-firstiter+I_exact,"-"],myfile)

n = np.array([2**i for i in range(0,8)])
error_vec = np.zeros(len(n))


for i in n[1:]:
    T_h = TrapezoidalRuleIntegration(i)
    eh = abs(T_h - I_exact) 
    np.append(error_vec,eh)
    eh2 = abs(firstiter - I_exact)
    firstiter = T_h
    WriteDataToFile([i,(xb-xa)/i,T_h,eh,eh2/eh],myfile)

    
myfile.close()


myfile = open("output_data\\SimpingForNami.csv", 'a') 
myfile.seek(0) 
myfile.truncate() 
WriteDataToFile(["n","h","EstimateIntegral","eh","e2h/eh"],myfile)

firstiter = SimpsonRuleIntegration(1)
WriteDataToFile([1,(xb-xa),firstiter,-firstiter+I_exact,"-"],myfile)

n = np.array([2**i for i in range(0,8)])
error_vec = np.zeros(len(n))


for i in n[1:]:
    T_h = SimpsonRuleIntegration(i)
    eh = abs(T_h - I_exact) 
    np.append(error_vec,eh)
    eh2 = abs(firstiter - I_exact)
    firstiter = T_h
    WriteDataToFile([i,(xb-xa)/i,T_h,eh,eh2/eh],myfile)

    
myfile.close()
```
