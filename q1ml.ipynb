```python
import numpy as  np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

plt.style.use('dark_background')

def WriteDataToFile(list,myfile):
    myfile.write(', '.join(str(item) for item in list)+'\n')
    
def da(t,c):
    return [-6*c[0],2*c[0],4*c[0]]


t_span = np.arange(0,1.05,.05)
sola = solve_ivp(da ,[0, 1],[1,0,0],t_eval = t_span)

myfile = open("output_data\\chemical_A.csv", 'a') 
myfile.seek(0) 
myfile.truncate() 
WriteDataToFile(["t_span","c_a","c_b","c_c"],myfile)
WriteDataToFile([0,1,0,0],myfile)
for i in range(len(sola.t)):
    WriteDataToFile([sola.t[i],sola.y[0][i],sola.y[1][i],sola.y[2][i]],myfile)
myfile.close()

fig, ax = plt.subplots(1, 1, figsize=(16, 9))
ax.plot(sola.t,sola.y[0],'-o',label = '$C_a$')
ax.plot(sola.t,sola.y[1],'-d',label = '$C_b$')
ax.plot(sola.t,sola.y[2],'-v',label = '$C_c$')
ax.legend(prop={'size': 16})
ax.set_title("Concentration vs time")
plt.show()
```
